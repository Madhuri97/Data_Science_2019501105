```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("rating_data.tsv", delimiter = '\t')
ratings.head(10)
```

```python
ratings = ratings[ratings["numVotes"] >= 10000]
ratings
```

```python
ratings.drop(["tconst"], axis = 1, inplace = True)
ratings
```

```python
ratings.to_csv("rankvsvotes.csv", index = False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(ratings["averageRating"], ratings["numVotes"], alpha=0.7, color = '#88c999')
plt.xlabel("votes")
plt.ylabel("rank")
plt.show()
```
